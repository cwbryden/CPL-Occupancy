```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Christopher\Jupyter Scripts\Data\CPL Occupancy Daily Data 2022 YTD.csv')
```

```python
df.head()
```

```python
df['Record Date'] = pd.to_datetime(df['Record Date'])
df1 = df[df['Record Date'].dt.month == 1]
```

```python
#df1.head()
```

```python
plt.title('CPL Occupancy by Day in January', fontdict = {'fontname': 'Serif', 'fontsize': 30})

plt.plot(df1['Day of Month'], df1['Avg. Occupancy'])
plt.plot(df1['Day of Month'], df1['Max. Occupancy'])
plt.plot(df1['Day of Month'], df1['Min. Occupancy'])

plt.xlabel('Day In January')
plt.ylabel('Occupancy')

plt.legend(['Avg.', 'Max', 'Min'])
plt.show()
```

#### Note that this data is not total number of patrons, but the average, minimum, and maximum number of patrons, given any 15 minute interval on a specific day of the month. Data pulled from vea.sensourceinc

### Creating another Scatter Plot using Searborn to view each data entry by day in each month.

```python
plt.title('CPL Occupancy by Day in Month YTD 2022', fontdict = {'fontname': 'Serif', 'fontsize': 30})
sns.scatterplot(data = df,
                x ='Day of Month', 
                y = 'Avg. Occupancy',
                hue = 'Month'
               );
plt.xlabel("Day of Month")
plt.ylabel("Occupancy")
plt.figure(dpi = 300)
```

## Total Number of Patrons Daily

Above we used the tdaily totals given our open hours and 15 minute increments from sensource to get an idea of patrons in the building at any given time. 

Here we will look at the total number of patrons that visit the Canton Public Library daily pulling our data from the 'Ins' column.

```python
plt.title('CPL Total # of Patrons by Day in Month YTD 2022', fontdict = {'fontname': 'Serif', 'fontsize': 30})
sns.scatterplot(data = df,
                x ='Day of Month', 
                y = 'Ins',
                hue = 'Month'
               );
plt.xlabel("Day of Month")
plt.ylabel("Total # of Patrons")
plt.figure(dpi = 300)
```

For viewing purposes instead of a scatter plot we can best visualize the trends with a line plot.

```python
plt.title('CPL Total # of Patrons by Day in Month YTD 2022', fontdict = {'fontname': 'Serif', 'fontsize': 15})

plt.plot(df1['Day of Month'], df1['Ins'])
plt.plot(df2['Day of Month'], df2['Ins'])
plt.plot(df3['Day of Month'], df3['Ins'])

plt.xlabel('Day of Month')
plt.ylabel('Total # of Patrons')

plt.legend(['January', 'February', 'March'])
plt.show()
```

## Box and Whisker Plot by Days of Week

Another visualization we can make with this data is a box and whisker plot to get an idea of the variance of our total number of patrons visiting based on day of the week

```python
daysData = sns.boxplot(data = df,
                      x = 'Day Of Week',
                      y = 'Ins',
                      palette = "Set2")

daysLabels = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

#daysData.set_xticklabels(daysData.get_xticklabels(), rotation = 90)

daysData.set_xticklabels(daysLabels, rotation = 90)
plt.title('CPL Total # of Patrons by Day of Week YTD 2022', fontdict = {'fontname': 'Serif', 'fontsize': 15})
plt.xlabel('Day of Week')
plt.ylabel('Total Number of Patrons')
plt.figure(figsize = (6, 4), dpi = 300)

```

## We can also make a Swarm Plot which is a Box and Whisker Plot that also shows our individual data points.

```python
daysData = sns.boxplot(data = df,
                      x = 'Day Of Week',
                      y = 'Ins',
                      palette = 'Set2')

daysData = sns.swarmplot(data = df,
                      x = 'Day Of Week',
                      y = 'Ins',
                      color = "0.25")


daysLabels = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

#daysData.set_xticklabels(daysData.get_xticklabels(), rotation = 90)

daysData.set_xticklabels(daysLabels, rotation = 90)
plt.title('CPL Total # of Patrons by Day of Week YTD 2022', fontdict = {'fontname': 'Serif', 'fontsize': 15})
plt.xlabel('Day of Week')
plt.ylabel('Total Number of Patrons')
plt.figure(figsize = (6, 4), dpi = 300)
```
